```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
# Loading the data
df = pd.read_csv("metadata.csv")
df
```

```python
# Obtener valores Ãºnicos de Product Type y Section
unique_product_types = df['Product Type'].unique()
unique_sections = df['Section'].unique()

print("Types of products:", unique_product_types)
print("Unique sections:", unique_sections)
```

```python
# Products visual classification
base_folder = '.'
combinations = df[['Product Type', 'Section']]
for _, row in combinations.iterrows():
    product_type = row['Product Type']
    section = row['Section']
    filtered_df = df[(df['Product Type'] == product_type) & (df['Section'] == section)]
    images = filtered_df['Path'].head(10)
    
    fig, axes = plt.subplots(1, len(images), figsize=(20, 5))
    fig.suptitle(f'Product Type: {product_type}, Section: {section}', fontsize=16)
    
    for ax, img_path in zip(axes, images):
        full_path = os.path.join(base_folder, img_path)
        img = plt.imread(full_path)
        ax.imshow(img)
        ax.axis('off')
        ax.set_title(os.path.basename(img_path))

    plt.show()
```
